```python
import pandas as pd
import numpy as np
from ggplot import *

data = pd.DataFrame.from_csv("train.csv")
data['GrLivArea'] = data['GrLivArea'].apply(np.log)
```

```python
p = ggplot(data,aes(y='SalePrice',x='GrLivArea'))
p + geom_point() + facet_wrap(x='Neighborhood')
```

```python
ggplot(data,aes(x='LotArea',y='SalePrice',color='Heating')) + geom_point() + scale_x_log()
```

```python
ggplot(data,aes(x='BldgType',y='SalePrice')) + geom_boxplot()

#print(data['SalePrice'][c=='Corner'].mean()/data['SalePrice'][c=='Inside'].mean())
```
