```python
import sklearn as sk
import sklearn.covariance as cov
import pandas as pd
import numpy as np
from ggplot import *

data = pd.DataFrame.from_csv("inflation.csv",index_col=None)
```

```python
def date_convert(date):
    year, month = date.split(sep='-')
    return float(year)+float(month)/12
```

```python
usa = data[data['LOCATION']=='USA']
usa = usa[usa['SUBJECT'] == 'TOT']
usa = usa[usa['TIME'] > '2005']
usa = usa[usa['MEASURE']=='AGRWTH']
usa = usa[usa['FREQUENCY']=='M']
usa["TIME"] = usa['TIME'].map(date_con)
```
